```python
def get_files(filepath):
    all_files = []
    for root, dirs, files in os.walk(filepath):
        files = glob.glob(os.path.join(root,'*.json'))
        for f in files :
            all_files.append(os.path.abspath(f))
    
    return all_files
```

```python
song_files = get_files('data/song_data/')
log_files = get_files('data/log_data')
```

```python
song=pd.DataFrame()

for f in song_files:
    song=song.append(pd.read_json(f,lines=True),ignore_index=True)
```

```python
log=pd.DataFrame()

for f in log_files:
    log=log.append(pd.read_json(f,lines=True),ignore_index=True)
    
# log
```

```python
pd.merge(song,log,left_on=['artist_name','title','duration'],right_on=['artist','song','length'], how="inner")
```
